```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats

# Sample dataset (must be positive values only)
data = {
    'Income': [25000, 30000, 27000, 45000, 52000, 28000, 26000, 47000, 100000, 29000]
}
df = pd.DataFrame(data)

# Histogram before transformation
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.histplot(df['Income'], kde=True, color='skyblue')
plt.title('Before Box-Cox Transformation')

# Apply Box-Cox transformation
df['Income_BoxCox'], fitted_lambda = stats.boxcox(df['Income'])

# Histogram after transformation
plt.subplot(1, 2, 2)
sns.histplot(df['Income_BoxCox'], kde=True, color='lightgreen')
plt.title('After Box-Cox Transformation')

plt.tight_layout()
plt.show()

# Display the lambda used
print(f"Optimal λ (lambda) for Box-Cox transformation: {fitted_lambda:.4f}")
```
