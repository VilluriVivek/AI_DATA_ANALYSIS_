```python
import pandas as pd
from scipy.stats.mstats import winsorize
import seaborn as sns
import matplotlib.pyplot as plt

# Sample dataset
data = {
    'Math_Score': [78, 85, 90, 95, 88, 76, 94, 98, 200, 77, 82, 86, 93, 75, 89]
}
df = pd.DataFrame(data)

# Before Winsorization
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.boxplot(y=df['Math_Score'], color='salmon')
plt.title('Before Winsorization')

# Apply Winsorization (limit 10% from both ends)
winsorized_scores = winsorize(df['Math_Score'], limits=[0.10, 0.10])
df['Winsorized_Score'] = winsorized_scores

# After Winsorization
plt.subplot(1, 2, 2)
sns.boxplot(y=df['Winsorized_Score'], color='lightgreen')
plt.title('After Winsorization')

plt.tight_layout()
plt.show()
```
